### Set
    - A set is an unordered collection of hashable objects. 
    - It is iterable, mutable and has unique elements
    - Defining set
        x = {'data', 'structure'}
    - Some of the function which you can perform on set. By considering x1 and x2 as two sets
        - Union
            - x1.union(x2)
            - Alterative way to define : x1 | x2
        - Intersection
            - x1.intersection(x2)
            - Alternative way to define: x1 && x2
        - Difference
            - x1.difference(x2)
            - Alternative way : x1 - x2
        - Symmetric Difference
            - x1.symmetric_difference(x2)
            - x1 ^ x2
        - Subset
            - x1.subset(x2)
            - x1 <= x2

### Named Tuple
    - The named tuple of collection provides an extensionn of the build in tuple data types. They are immutable
    - <strong> They are mainly useful when several tuples are used in an application and it is important to keep track of each of the tuples in terms of what they represent </strong>
    - Sytax:
        nt = namedtuple(typename, field_names)
    - Example:
        from collections import namedtuple
        Book = namedtuple('Book',['name','ISBN','Quantity'])
        # Using named tuple
        Book1 = Book('Hands on DS', '134342432423', '5')
        print('Using index ISBN:' + Book1[1])
        print('Using key ISBN:' + Book1.ISBN)

### deque
    - Deque is the double-ended-queue
    - It supports pop and appened elemet from both sides of the list
    - Deque are implemented as double-linked list, which are very efficient for inserting and deleting elemet inn O(1)times.
    - Example:
        from collections import namedtuple
        s = deque()   # Creates an empty deque
        print(s)
        my_queue = deque([1, 2, 'Name'])
        print(my_queue)
    - Functions of deque
        - myqueue.appened(age) - This will appened age at the right end of the list
        - myqueue.appendleft(age) - This will append age at the left end of the list
        - myqueue.pop() - Delete the right most value
        - myqueue.popleft() - Delete the left most value

### defaultdict
    - The default dictionary(defaultdict) is a subclass of the build in dictionary class(dict) that has the same method and operation as the dictionary class, with the only difference being that it is never raises a KeyError as normal dictionary would. 
    - DefaultDict is the convenient way to initialize dictionaries
    - Syntax:
        d = defaultdict(def_values)
    - Example:
        from collections import defaultdict
        dd = defaultdict(int)
        words = str.split('data python data data structure data python')
        for word in words:
            dd[word] += 1
        print(dd)

### ChainMap
    - It is used to create list of dictionaries.
    - The collection.ChainMap data structure combinnes several dictionaries into a single mapping
    - Whenever a key is searched i the chainmap, it looks through all the dictionaries one by one, until the key is not found.
    - Example:
        from collections import ChainMap
        dict1 = {"data": 1, "structure": 2}
        dict2 = {"python": 3, "language": 4}
        chain = ChainMap(dict1, dict2)
        print(chain)
        print(list(chain.keys()))
        print(list(chain.values()))
        print(chain["data"])
        print(chain["language"])

### Counter
    - The hashtable is the object whose hash value will remain the same during its lifetime in the program. The counter is use to count the number of hashtable objects
    - For example dictionary key is the hash objects and corresponding value is the count of the object. 
    - In other words, 'counter' object create a hash table in which the element and their count are stored as dictionary keys and value pairs.
    - Example: 
        from collections import Counter
        inventory = Counter('hello')
        print(inventory)
        print(inventory['l'])
        print(inventory['e'])
        print(inventory['o'])

### UserDict,Userist,UserString
#### UserDict
    - UserDict wraps the dict objects. We can add customized function to the dictionary
    - This is very useful for application where we want to add/update/modify the functionalities of the dictionary. 
    - Example: Where insert option is disabled
        # we can not push to this user dictionary
        from collections import UserDict
        class MyDict(UserDict):
            def push(self, key, value): 
                raise RuntimeError("Cannot insert")
        d = MyDict({'ab':1, 'bc': 2, 'cd': 3})
        d.push('b', 2)
#### Userlist
    - This is similar to user dict but it wraps list objects
    - Example: where insert option is disabled
        from collections import UserList
        class MyList(UserList):
            def push(self, key):
                raise RuntimeError("Cannot insert in the list")
        d = MyList([11, 12, 13])
        d.push(2)
#### UserString
    - Strings can be considered as an array of characters. In Python, a character is a string of one length and acts as a container that wraps a string object. It can be used to create strings with customized functionalities.
    - Example: Appened option is disabled
        from collections import UserString
        class MyString(UserString):
            def append(self, value):
                self.data += value
        s1 = MyString("data")
        print("Original:", s1)
        s1.append('h')
        print("After append: ", s1)
